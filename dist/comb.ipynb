```python
import csv
import json

# Read the JSON data from sec_tickers_ciks.json
with open("sec_tickers_ciks.json", "r") as f:
    json_data = json.load(f)

# Create a dictionary for easy lookup of CIKs
cik_to_title = {}
for item in json_data.values():
    cik_to_title[str(item['cik_str'])] = item['title']

# Read the CSV data from MISSING_FIRMS_FINAL.csv and write to a new CSV file
with open("MISSING_FIRMS_FINAL.csv", "r") as csv_file:
    csv_reader = csv.reader(csv_file)
    header = next(csv_reader)  # Skip the header row

    with open("NEW_MISSING_FIRMS_FINAL.csv", "w", newline='') as new_csv_file:
        csv_writer = csv.writer(new_csv_file)
        
        # Write the new header
        csv_writer.writerow(["Symbol", "CIK", "Title"])

        # Iterate through rows to write new data
        for row in csv_reader:
            symbol = row[0]
            cik = row[1]
            title = cik_to_title.get(cik, "Title Not Found")
            csv_writer.writerow([symbol, cik, title])
```

```python
import csv

# Define the new headers in all caps
new_headers = [
    "DATE",
    "NAME",
    "CIK",
    "FIRM",
    "SOURCE",
    "BLACK",
    "MALE",
    "FEMALE",
    "LGBT",
    "NON_BINARY",
    "ASIAN",
    "LATINX",
    "DIRECTORS",
    "NOTES",
    "DND_GENDER",
    "DND_DEMO",
]

# Open the CSV file and create a TSV writer
with open("new_missing_firms_final.csv", "r") as csv_file, open(
    "converted_firms_final.tsv", "w", newline=""
) as tsv_file:

    csv_reader = csv.DictReader(csv_file)
    tsv_writer = csv.DictWriter(tsv_file, fieldnames=new_headers, delimiter="\t")

    # Write the new headers to the TSV file
    tsv_writer.writeheader()

    # Iterate through each row in the CSV file and write it to the TSV file
    for row in csv_reader:
        new_row = {
            "DATE": "2022",
            "NAME": row["Title"],
            "CIK": row["CIK"],
            "FIRM": row["Symbol"],
            "SOURCE": "proxy",
            "BLACK": "",
            "MALE": "",
            "FEMALE": "",
            "LGBT": "",
            "NON_BINARY": "",
            "ASIAN": "",
            "LATINX": "",
            "DIRECTORS": "",
            "NOTES": "",
            "DND_GENDER": "",
            "DND_DEMO": "",
        }
        tsv_writer.writerow(new_row)
```
